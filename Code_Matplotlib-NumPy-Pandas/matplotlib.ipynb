```python
# 数据在x轴的位置 
x = range(2, 26, 2)
# 数据在y轴的位置
y = [11, 23, 11.3, 35, 45, 65, 35, 27, 76, 8, 11, 45]
```

```python
# 传入x y 通过plot绘制出折线图
plt.plot(x, y)
# 展示图表
plt.show()
```

```python
fig = plt.figure(figsize=(10,5),dpi=100)
plt.plot(x,y)
plt.show()

# figure图形图标的意思，在这里指的就是我们要画的图
# 通过实例化一个figure并且传递参数，能够在后台自动使用该figure实例
# figsize 用来设置图像大小
# 在图像模糊的时候 可以传入dpi参数来让图片更加清晰
# savefig 保存图片
# 参数为 图像名称以及格式
# 可以保存为svg矢量图格式，放大不会有锯齿
# plt.savefig("name.png")
```

### xticks(ticks, [labels], **kwargs)

举个例子：
**plt.xticks(x,calendar.month_name[1:13],color='blue',rotation=60)  #参数x空值X轴的间隔，第二个参数控制每个间隔显示的文本，后面两个参数控制标签的颜色和旋转角度**

星号：https://www.cnblogs.com/empty16/p/6229538.html

```python
from matplotlib import pyplot as plt
import random
x = range(0, 60)
y = [random.randint(10, 40) for i in range(60)]
plt.plot(x, y)
# 调整x轴刻度
_x = x[::5]
_xtick_labels = ["{}分钟".format(i) for i in _x]
# plt.xticks(_x, _xtick_labels)
# rotation 刻度旋转
plt.xticks(_x, _xtick_labels, rotation=50, fontproperties="STSong")
plt.show()
```

```python
from matplotlib import pyplot as plt
import random
# 设置字体
import matplotlib
font = {"family":"MicroSoft YaHei",
       "weight":"bold",
       "size":"12"}
matplotlib.rc("font", **font)

x = range(0, 60)
y = [random.randint(10, 40) for i in range(60)]

plt.plot(x, y)
_x = x[::5]
_xtick_labels = ["{}分钟".format(i) for i in _x]
plt.xticks(_x, _xtick_labels, rotation=50)
# 添加x轴描述信息
plt.xlabel("时间")
# 设置y轴表述信息
plt.ylabel("温度")
# 设置标题
plt.title("每分钟气温变化图")
plt.show()
```

```python

font = {"family":"MicroSoft YaHei",
       "weight":"bold",
       "size":"12"}
matplotlib.rc("font", **font)
x = range(0, 60)
y = [random.randint(10, 40) for i in range(60)]

plt.plot(x, y)
_x = x[::5]
_xtick_labels = ["{}分钟".format(i) for i in _x]
plt.xticks(_x, _xtick_labels, rotation=50)
plt.xlabel("时间")
plt.ylabel("温度")
plt.title("每分钟气温变化")
# 绘制网格
# plt.grid()
# alpha 设置透明度
plt.grid(alpha=0.4)
plt.show()
```

```python

plt.plot(x, y,
        color="red",# 线条颜色
        linestyle="--",# 线条风格
        linewidth=2,# 线条粗细
        alpha=0.5 #透明度
        )
_x = x[::5]
_xtick_labels = ["{}min分钟".format(i) for i in _x]
plt.xticks(_x, _xtick_labels, rotation=50)
plt.xlabel("时间")
plt.ylabel("温度")
plt.title("每分钟气温变化")
plt.grid(alpha=0.4)
plt.show()
```
